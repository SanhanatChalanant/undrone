```python
#เอาก่อนหน้านี้ทั้งหมดมารวมกันเป็นloop
for idx,row in splitundrone.iterrows() :
    disfirst=[]
    dislast=[]
    first,last=splitundrone['geometry'][idx].boundary#หาจุดเริ่มกับจุดท้าย
    nearpoint1=nearest_points(first,pointstation)#หาจุดที่ใกล้
    nearpoint2=nearest_points(last,pointstation)#หาจุดที่ใกล้
    disfirst.append(first)#เอาไปใส่ใน[]
    disfirst.append(nearpoint1[1])#เอาไปใส่ใน[]
    dislast.append(last)#เอาไปใส่ใน[]
    dislast.append(nearpoint2[1])#เอาไปใส่ใน[]
    distancefirst=gpd.GeoDataFrame(geometry=disfirst,crs=32647)#เอาไปเก็ยในdataframe
    distancelast=gpd.GeoDataFrame(geometry=dislast,crs=32647)#เอาไปเก็ยในdataframe
    dis1=distancefirst.distance(distancefirst.shift())#หาระยะทาง
    dis2=distancelast.distance(distancelast.shift())#หาระยะทาง
    if dis1[1]<dis2[1] :#เอาตัวที่ระยะน้อยกว่า
        near.append(nearpoint1[1])
    elif dis2[1]<=dis1[1] :#เอาตัวที่ระยะน้อยกว่า
        near.append(nearpoint2[1])
    else :
        print('1')
    nearstation = gpd.GeoDataFrame(geometry=near,crs=32647)#ใส่พิกัดด้วย
    rows=[0,1]
    distancefirst.drop(rows,axis=0)#clearlist
    distancelast.drop(rows,axis=0)#clearlist
    
```

```python
near=[]
azi=[]
direct=[]
for idx,row in splitundrone.iterrows() :
    disfirst=[]
    dislast=[]
    first,last=splitundrone['geometry'][idx].boundary#หาจุดเริ่มกับจุดท้าย
    nearpoint1=nearest_points(first,pointstation)#หาจุดที่ใกล้
    nearpoint2=nearest_points(last,pointstation)#หาจุดที่ใกล้
    disfirst.append(first)#เอาไปใส่ใน[]
    disfirst.append(nearpoint1[1])#เอาไปใส่ใน[]
    dislast.append(last)#เอาไปใส่ใน[]
    dislast.append(nearpoint2[1])#เอาไปใส่ใน[]
    distancefirst=gpd.GeoDataFrame(geometry=disfirst,crs=32647)#เอาไปเก็ยในdataframe
    distancelast=gpd.GeoDataFrame(geometry=dislast,crs=32647)#เอาไปเก็ยในdataframe
    dis1=distancefirst.distance(distancefirst.shift())#หาระยะทาง
    dis2=distancelast.distance(distancelast.shift())#หาระยะทาง
    if dis1[1]<dis2[1] :#เอาตัวที่ระยะน้อยกว่า
        near.append(nearpoint1[1])
        az=math.degrees(math.atan2((nearpoint1[1].xy[0][0]-first.xy[0][0]),(nearpoint1[1].xy[1][0]-first.xy[1][0])))
        if az<0 :
            az+=360
        azi.append(az)
        if az>0 and az<180 :
            direct.append('right from ct')
        elif az>180 and az<360 :
            direct.append('left from ct')
    elif dis2[1]<=dis1[1] :#เอาตัวที่ระยะน้อยกว่า
        near.append(nearpoint2[1])
        az=math.degrees(math.atan2((nearpoint2[1].xy[0][0]-last.xy[0][0]),(nearpoint2[1].xy[1][0]-last.xy[1][0])))
        if az<0 :
            az+=360
        azi.append(az)
        if az>0 and az<180 :
            direct.append('right from ct')
        elif az>180 and az<360 :
            direct.append('left from ct')
    else :
        print('1')
    #nearstation = gpd.GeoDataFrame(geometry=near,crs=32647)#ใส่พิกัดด้วย
    splitundrone['azimuth'] = pd.DataFrame(azi)
    splitundrone['direction'] = pd.DataFrame(direct)
    rows=[0,1]
    distancefirst.drop(rows,axis=0)#clearlist
    distancelast.drop(rows,axis=0)#clearlist
```
